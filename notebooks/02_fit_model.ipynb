# 02 · Fit the Bayesian double-Poisson model
Fit via MCMC (PyMC), check convergence (R-hat, ESS, divergences), and run a demo prediction.

```python
import sys; sys.path.insert(0, '../src')
from footpred import data, model, predict
df = data.load_processed()
idata = model.fit(df, draws=1000, tune=1000, chains=4)
model.save_trace(idata)
```

```python
print(model.convergence_summary(idata))
print('divergences:', model.n_divergences(idata))
```

```python
import arviz as az
az.plot_trace(idata, var_names=['intercept','home_adv','sd_att','sd_def'])
```

```python
# Demo: reproduce a reel-style single prediction
pred = predict.predict(idata, 'Brazil', 'Haiti', neutral=True)
pred.as_dict()
```

```python
# Top team attack strengths (posterior mean)
import numpy as np, pandas as pd
atk = idata.posterior['atk'].mean(('chain','draw')).values
teams = list(idata.attrs['teams'])
pd.Series(atk, index=teams).sort_values(ascending=False).head(15)
```
