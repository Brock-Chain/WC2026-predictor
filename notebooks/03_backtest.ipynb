# 03 · Temporal backtest
Expanding-window, no-leakage backtest. Compare the Bayesian model vs Elo on RPS / log-loss / Brier + calibration. Bayesian refits per fold (slow) — keep folds few.

```python
import sys; sys.path.insert(0, '../src')
from footpred import data, backtest
df = data.load_processed()
# Score the most recent ~2 years in yearly folds.
folds = ['2023-01-01', '2024-01-01']
```

```python
elo = backtest.temporal_backtest(df, backtest.elo_factory(), folds)
print('Elo:', elo.summary())
```

```python
bayes = backtest.temporal_backtest(df, backtest.bayes_factory(draws=500, tune=500, chains=2), folds)
print('Bayes:', bayes.summary())
```

```python
# Calibration of the Bayesian home-win probability
backtest.calibration_table(bayes.per_match)
```

**Decide here, empirically:** if low-score calibration is off → add Dixon–Coles; if recent form is underweighted → add time-decay weighting; if friendlies hurt RPS → set `competitive_only=True` in the data step.
